## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
corrmat=pandas_df.corr()
fig,ax=plt.subplots(figsize=(21,14))
sns.heatmap(corrmat,annot=True,annot_kws={'size':12})
```

```
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
     for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr

corr_features = correlation(pandas_df, 0.5)
len(set(corr_features))
```

```
pandas_df = pandas_df.drop(columns=["fare_amount","tip_amount","total_amount"])
pandas_df.head()
```

```
pandas_df_base=pandas_df.drop(columns=["hour_of_day","day_of_week","day","month","dropoff_location_id","pickup_location_id","imp_surcharge","extra","payment_type","rate_code","mta_tax"])
pandas_df_base.head()
```

```
pandas_df_base=pandas_df_base.drop(index=pandas_df_base[pandas_df_base["trip_duration"]>16564].index)
pandas_df_base=pandas_df_base.drop(index=pandas_df_base[pandas_df_base["trip_distance"]>22].index)
pandas_df_base=pandas_df_base.drop(index=pandas_df_base[pandas_df_base["tolls_amount"]>12.86].index)
pandas_df_base=pandas_df_base.drop(columns=["_c0"])
pandas_df_base.head()
```

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(pandas_df_base)
scaled_features=scaler.transform(pandas_df_base)
```

```
import pandas as pd
pandas_df_base=pd.DataFrame(scaled_features, columns=pandas_df_base.columns)
display(pandas_df_base)
```

```
import pandas as pd
result_df_base=pd.DataFrame()
import sklearn.metrics as metrics
import sklearn.cluster as cluster
for i in range(3,20):
    labels=cluster.KMeans(n_clusters=i,init="k-means++",random_state=200).fit(pandas_df_base).labels_
    sil=metrics.silhouette_score(pandas_df_base,labels,metric="euclidean",sample_size=1000,random_state=200)
    from sklearn.cluster import KMeans
    model = KMeans(n_clusters=i)
    predicted=model.fit_predict(pandas_df_base)
    pandas_df_base["cluster"]=predicted
    std=pandas_df_base["cluster"].value_counts().std()
    print ("Silhouette score for k(clusters) = "+str(i)+" is " +str(sil))
    this_result_df=pd.DataFrame({"k":i,"sil":sil,"std":std},index=[0])
    result_df_base=result_df_base.append(this_result_df)    

number1=result_df_base.assign(experiment_name="base")
number1
```

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.scatter(data=result_df_base, x="k", y="sil")
sns.lineplot(data=result_df_base, x="k", y="sil")
plt.title("base")
plt.xlabel("number of cluster")
plt.ylabel("sil_score")
plt.show()
```
